```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
f = open('PT.txt', 'r')
data = f.read()
nums = list(map(float,data.split()))
f.close()
x = np.zeros(0)
y = np.zeros(0)
for i in range(len(nums)): 
    if i % 2 == 0: 
        y = np.append(y, nums[i])
    else :
        x = np.append(x, nums[i])
plt.figure(figsize=(10, 8))
plt.xlabel('P')
plt.ylabel('T')
plt.grid(True)
plt.title('T(P) for N = 10000000, M = 1250000')
plt.plot(y, x)
plt.plot(y, x, 'ro')
plt.plot(y, [1.34923, 1.34923, 1.34923, 1.34923], label = 'qsort')
plt.legend()
plt.show()
```
